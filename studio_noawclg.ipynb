```python
import noawclg.main as main
from noawclg.main import get_noaa_data as gnd
from noawclg.plot import plot_data_from_place as pdp
import matplotlib.pyplot as plt


date_base = '21/05/2023'

main.set_date(date_base)
data_noaa = gnd()


key = data_noaa.get_noaa_keys()
list_keys = [k for k in list(key.keys())][4:]
print(list_keys)
```

```python

place = 'juazeiro BA'

jua_pet = pdp(place=place,data=data_noaa)

i=0
size=len(list_keys)
for k in list_keys:
    i += 1
    jua_pet.path_file = f'keys_plots/plot_{k}.png'

    jua_pet.key_noaa = k
    jua_pet.title = key[k]
    print(f'rendering {i}/{size} ...')
    jua_pet.show=False
    jua_pet.render()
    
```
